```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy.stats import expon

title = 'Exponential distribution - PDF'
# title = 'Exponential distribution - CDF'
fig, ax = plt.subplots()
t = np.linspace(0, 20, 1000)
parameters_l = [0.5, 1, 2, 5, 10]

for l in parameters_l:
    if title == 'Exponential distribution - PDF':
        ax.plot(t, expon.pdf(t,scale=1/l), linewidth=3, label='$\lambda=$'+str(l))
    if title == 'Exponential distribution - CDF':
        ax.plot(t, expon.cdf(t,scale=1/l), linewidth=3, label='$\lambda=$'+str(l))

ax.set_xlabel('$t$', fontsize=15)
ax.set_ylabel('$f_{\lambda}(t)$', fontsize=15)
ax.set_title(title, fontsize=15)
ax.legend(loc='best', frameon=True)
ax.grid(True)
ax.set_ylim(-0.1,1.1)
```

```python
from scipy.stats import gamma

title = 'Gamma distribution - PDF'
# title = 'Gamma distribution - CDF'
fig, ax = plt.subplots()
t = np.linspace(0, 20, 1000)
parameters_k = [0.5, 1, 2, 3, 5, 10]
parameters_l = [1]

for l in parameters_l:
    for k in parameters_k:
        if title == 'Gamma distribution - PDF':
            ax.plot(t, gamma.pdf(t,k,scale=1/l), linewidth=3, label='$k=$'+str(k)+'$, \lambda=$'+str(l))
        if title == 'Gamma distribution - CDF':
            ax.plot(t, gamma.cdf(t,k,scale=1/l), linewidth=3, label='$k=$'+str(k)+'$, \lambda=$'+str(l))

ax.set_xlabel('$t$', fontsize=15)
ax.set_ylabel('$f_{k, \lambda}(t)$', fontsize=15)
ax.set_title(title, fontsize=15)
ax.legend(loc='best', frameon=True)
ax.grid(True)
ax.set_ylim(-0.1,1.1)
```

```python
from scipy.special import gamma

title = 'Gamma function'
fig, ax = plt.subplots()
t = np.linspace(0, 5, 1000)
ax.plot(t, gamma(t), linewidth=3)
ax.set_xlabel('$t$', fontsize=15)
ax.set_ylabel('$\Gamma(t)$', fontsize=15)
ax.set_title(title, fontsize=15)
ax.grid(True)
ax.set_ylim(0,30)
```
