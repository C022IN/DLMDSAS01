```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

n=10
x = uniform.rvs(scale=10, size=n) # original sample
y = 10*uniform.rvs(scale=10, size=n) # rescaled sample
p = (1/n)*np.ones(n)
```

In *bar diagrams* the *height* of the bars represents the probabilities:

```python
plt.bar(x, p, width = 0.05)
plt.title('Bar diagram - original sample', fontsize=15)
plt.xlabel('Value', fontsize=13)
plt.ylabel('Probability', fontsize=13)
plt.show()
plt.bar(y, p, width = 0.5)
plt.title('Bar diagram - rescaled sample', fontsize=15)
plt.xticks(color='r')
plt.xlabel('Value', fontsize=13, color='r')
plt.ylabel('Probability', fontsize=13)
plt.show()
```

In *histograms* the *area* of the rectangles represents the probabilities:

```python
equidistant_bins_x = np.arange(11)
equidistant_bins_y = tuple(10*np.array(equidistant_bins_x))

plt.hist(x, bins=equidistant_bins_x, density=True)
plt.title('Histogram (equidistant bins) - original sample', fontsize=15)
plt.xlabel('Value', fontsize=13)
plt.ylabel('Probability density', fontsize=13)
plt.show()
plt.hist(y, bins=equidistant_bins_y, density=True)
plt.title('Histogram (equidistant bins) - rescaled sample', fontsize=15)
plt.xticks(color='r')
plt.xlabel('Value', fontsize=13, color='r')
plt.yticks(color='r')
plt.ylabel('Probability density', fontsize=13, color='r')
plt.show()
```

```python
custom_bins_x = [0, 2, 5, 10]
custom_bins_y = tuple(10*np.array(custom_bins_x))

plt.hist(x, bins=custom_bins_x, density=True)
plt.title('Histogram (custom bins) - original sample', fontsize=15)
plt.xlabel('Value', fontsize=13)
plt.ylabel('Probability density', fontsize=13)
plt.show()
plt.hist(y, bins=custom_bins_y, density=True)
plt.title('Histogram (custom bins) - rescaled sample', fontsize=15)
plt.xticks(color='r')
plt.xlabel('Value', fontsize=13, color='r')
plt.yticks(color='r')
plt.ylabel('Probability density', fontsize=13, color='r')
plt.show()
```
