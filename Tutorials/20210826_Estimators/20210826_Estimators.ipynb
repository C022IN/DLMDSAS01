```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

theta = 11
n = 10
x = uniform.rvs(scale=theta, size=n)
y = np.zeros(n)
t1 = 2*np.mean(x)
t2 = np.max(x)
t3 = ((n+1)/n)*np.max(x)

fig, ax = plt.subplots()
plt.axhline(y=0, color='0.2', linestyle='-')
ax.scatter(x, y, s=100)
ax.get_yaxis().set_visible(False)
ax.set_xlim(-0.2,np.max([t1,t2,t3])+1.5)
ax.set_ylim(-0.15,0.15)
ax.set_title('Uniform sample from $[0, \\theta]$', fontsize=15)
fig.set_figwidth(8)
fig.set_figheight(3)
plt.show()

fig, ax = plt.subplots()
plt.axhline(y=0, color='0.2', linestyle='-')
ax.scatter(x, y, s=100)
ax.vlines(t1, -0.1, 0.1, colors='r', label='$\\bar T_n = \\frac{2}{n} \\sum_{i=1}^n X_i$')
ax.vlines(t2, -0.1, 0.1, colors='g', label='$T_n^{MLE} = \\max \{X_1, ..., X_n\}$')
ax.vlines(t3, -0.1, 0.1, colors='b', label='$T_n^* = \\frac{n+1}{n} T_n^{MLE}$')
ax.get_yaxis().set_visible(False)
ax.set_xlim(-0.2,np.max([t1,t2,t3])+1.5)
ax.set_ylim(-0.15,0.25)
ax.set_title('Estimators for $\\theta$ given uniform sample from $[0, \\theta]$', fontsize=15)
ax.legend(loc='upper center', ncol=3, fontsize='large', frameon=True)
fig.set_figwidth(8)
fig.set_figheight(4)
plt.show()
```
