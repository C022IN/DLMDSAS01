```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

#measurements = stats.norm.rvs(loc=20, scale=5, size=1000)
#measurements = stats.t.rvs(df=3, loc=20, scale=5, size=1000)
measurements = stats.expon.rvs(scale=5, size=1000)
mean = np.mean(measurements) 
std = np.std(measurements, ddof=1)
```

```python
# Q-Q plot

stats.probplot(measurements, dist="norm", plot=plt)
plt.show()
```

```python
# Histogram

plt.hist(measurements, color='b', bins=25)
plt.xlabel('Values')
plt.ylabel('Frequency in sample')
plt.show()

print('Skewness')
print(stats.skew(measurements))
print('Excess kurtosis')
print(stats.kurtosis(measurements))
```

```python
# Empirical and theoretical cumulative distribution function

x1 = np.sort(measurements)
n = x1.size
y1 = np.arange(1, n+1)/n
x2 = np.linspace(x1[0], x1[n-1])

fig, ax = plt.subplots()
ax.scatter(x=x1, y=y1, color='b')
ax.plot(x2, stats.norm.cdf(x2,loc=mean,scale=std), linewidth=2, color='r')
ax.set_xlabel('Values')
ax.set_ylabel('Cumulative distribution function')
plt.show()
```
