```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm


x = np.linspace(-4, 4, 1000)
mu = 0
sigma = 1
mean, var, skew, kurt = norm.stats(loc=mu, scale=sigma, moments='mvsk')
std = norm.std(loc=mu, scale=sigma)
median = norm.median(loc=mu, scale=sigma)
quartile1 = norm.ppf(0.25, loc=mu, scale=sigma)
quartile2 = norm.ppf(0.5, loc=mu, scale=sigma)
quartile3 = norm.ppf(0.75, loc=mu, scale=sigma)
mode = mu

fig, ax = plt.subplots()
ax.plot(x, norm.pdf(x, loc=mu, scale=sigma), label='pdf', c='tab:blue', linewidth=3)
plt.axvline(mean, label='mean', c='tab:brown')
plt.axvline(median, label='median', c='tab:green')
plt.axvline(quartile1, label='quartiles', c='tab:green', linestyle='dotted')
plt.axvline(quartile3, c='tab:green', linestyle='dotted')
ax.hlines(0.15, mean, mean+std, label='mean +/- std', colors='tab:red')
ax.hlines(0.15, mean, mean-std,colors='tab:red')
ax.vlines(mean-std, 0.14, 0.16, colors='tab:red')
ax.vlines(mean, 0.14, 0.16, colors='tab:red', linewidth=3)
ax.vlines(mean+std, 0.14, 0.16, colors='tab:red')
ax.scatter(mode, norm.pdf(mode, loc=mu, scale=sigma), label='mode', c='tab:gray', s=100)
ax.set_xlabel('$x$', fontsize=15)
ax.set_ylabel('$f_{\mu, \sigma}(x)$', fontsize=15)
ax.set_title('Normal distribution for $\mu=$'+str(mu)+', $\sigma=$'+str(sigma), fontsize=15)
ax.legend(loc='best', frameon=True)
ax.grid(True)
plt.show()

print('skewness: '+str(skew))
print('excess kurtosis: '+str(kurt))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma


x = np.linspace(0, 8, 1000)
k = 2
l = 1
mean, var, skew, kurt = gamma.stats(k, scale=1/l, moments='mvsk')
std = gamma.std(k, scale=1/l)
median = gamma.median(k, scale=1/l)
quartile1 = gamma.ppf(0.25, k, scale=1/l)
quartile2 = gamma.ppf(0.5, k, scale=1/l)
quartile3 = gamma.ppf(0.75, k, scale=1/l)
mode = (k-1)/l

fig, ax = plt.subplots()
ax.plot(x, gamma.pdf(x, k, scale=1/l), label='pdf', c='tab:blue', linewidth=3)
plt.axvline(mean, label='mean', c='tab:brown')
plt.axvline(median, label='median', c='tab:green')
plt.axvline(quartile1, label='quartiles', c='tab:green', linestyle='dotted')
plt.axvline(quartile3, c='tab:green', linestyle='dotted')
ax.hlines(0.15, mean, mean+std, label='mean +/- std', colors='tab:red')
ax.hlines(0.15, mean, mean-std,colors='tab:red')
ax.vlines(mean-std, 0.14, 0.16, colors='tab:red')
ax.vlines(mean, 0.14, 0.16, colors='tab:red', linewidth=3)
ax.vlines(mean+std, 0.14, 0.16, colors='tab:red')
ax.scatter(mode, gamma.pdf(mode, k, scale=1/l), label='mode', c='tab:gray', s=100)
ax.set_xlabel('$x$', fontsize=15)
ax.set_ylabel('$f_{k, \lambda}(x)$', fontsize=15)
ax.set_title('Gamma distribution for $k=$'+str(k)+', $\lambda=$'+str(l), fontsize=15)
ax.legend(loc='best', frameon=True)
ax.grid(True)
plt.show()

print('skewness: '+str(skew))
print('excess kurtosis: '+str(kurt))
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.stats import binom


n = 10
p = 0.333
x = np.arange(n+1)
mean, var, skew, kurt = binom.stats(n, p, moments='mvsk')
std = binom.std(n, p)
median = binom.median(n, p)
quartile1 = binom.ppf(0.25, n, p)
quartile2 = binom.ppf(0.5, n, p)
quartile3 = binom.ppf(0.75, n, p)
mode = math.floor((n+1)*p)

fig, ax = plt.subplots()
plt.bar(x, binom.pmf(x, n, p), label='pmf', color='tab:blue', width=0.4)
plt.axvline(mean, label='mean', c='tab:brown')
plt.axvline(median, label='median', c='tab:green')
plt.axvline(quartile1, label='quartiles', c='tab:green', linestyle='dotted')
plt.axvline(quartile3, c='tab:green', linestyle='dotted')
ax.hlines(0.15, mean, mean+std, label='mean +/- std', colors='tab:red')
ax.hlines(0.15, mean, mean-std,colors='tab:red')
ax.vlines(mean-std, 0.14, 0.16, colors='tab:red')
ax.vlines(mean, 0.14, 0.16, colors='tab:red', linewidth=3)
ax.vlines(mean+std, 0.14, 0.16, colors='tab:red')
ax.scatter(mode, binom.pmf(mode, n, p), label='mode', c='tab:gray', s=100)
ax.set_xlabel('$x$', fontsize=15)
ax.set_ylabel('$f_{n, p}(x)$', fontsize=15)
ax.set_title('Binomial distribution for $n=$'+str(n)+', $p=$'+str(p), fontsize=15)
ax.legend(loc='best', frameon=True)
ax.grid(True)
plt.show()

print('skewness: '+str(skew))
print('excess kurtosis: '+str(kurt))
```
