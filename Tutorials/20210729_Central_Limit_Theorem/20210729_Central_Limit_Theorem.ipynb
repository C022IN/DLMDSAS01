```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

title = 'Normal distribution - PDF'
# title = 'Normal distribution - CDF'
fig, ax = plt.subplots()
t = np.linspace(-10, 10, 1000)
parameters_mu = [0, 5]
parameters_sigma = [1, 2]

for mu in parameters_mu:
    for sigma in parameters_sigma:
        if title == 'Normal distribution - PDF':
            ax.plot(t, norm.pdf(t,loc=mu,scale=sigma), linewidth=3, label='$\mu=$'+str(mu)+'$, \sigma^2=$'+str(sigma*sigma))
        if title == 'Normal distribution - CDF':
            ax.plot(t, norm.cdf(t,loc=mu,scale=sigma), linewidth=3, label='$\mu=$'+str(mu)+'$, \sigma^2=$'+str(sigma*sigma))

ax.set_xlabel('$t$', fontsize=15)
if title == 'Normal distribution - PDF':
    ax.set_ylabel('$\phi_{\mu, \sigma^2}(t)$', fontsize=15)
if title == 'Normal distribution - CDF':
    ax.set_ylabel('$\Phi_{\mu, \sigma^2}(t)$', fontsize=15)
ax.set_title(title, fontsize=15)
ax.legend(loc='best', frameon=True)
ax.grid(True)
```

```python
import numpy as np
import random
import matplotlib.pyplot as plt
from scipy.stats import gamma
from scipy.stats import expon
from scipy.stats import uniform

N = 50000 # sample size of Monte-Carlo simulation
n = 500 # number of i.i.d. random variables in each sample
def ppf(q): # percent point functions (quantiles) of examplary distributions, e.g.
    return gamma.ppf(q,2) # Gamma distribution
    #return expon.ppf(q) # exponential distribution
    #return uniform.ppf(q) # uniform distribution

sample_sum = [0]*N
q = np.random.rand(N,n)
sample = ppf(q)
for I in range(0, N):
    sample_sum[I] = sum(sample[I])
        
fig, ax = plt.subplots()
ax.hist(sample_sum, 100, density=True)
ax.set_xlabel('Unscaled sum $\sum_{i=1}^n x_i$ of $n$ i.i.d. random numbers', fontsize=15)
ax.set_ylabel('Relative frequency in $N$ samples of $\sum_{i=1}^n x_i$', fontsize=15)
ax.set_title('$N$ = '+str(N)+' samples with $n$ = '+str(n)+' i.i.d. random numbers', fontsize=15)
fig.set_figwidth(10)
fig.set_figheight(10)
plt.show()
```
