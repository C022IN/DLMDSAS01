```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

title = 'Chi squared distribution - PDF'
# title = 'Chi squared distribution - CDF'
fig, ax = plt.subplots()
t = np.linspace(0.01, 20, 1000)
parameters_n = [2, 5, 10, 20]

for n in parameters_n:
    if title == 'Chi squared distribution - PDF':
        ax.plot(t, chi2.pdf(t,n), linewidth=3, label='$n=$'+str(n))
    if title == 'Chi squared distribution - CDF':
        ax.plot(t, chi2.cdf(t,n), linewidth=3, label='$n=$'+str(n))

ax.set_xlabel('$t$', fontsize=15)
if title == 'Chi squared distribution - PDF':
    ax.set_ylabel('$f_{n}(t)$', fontsize=15)
if title == 'Chi squared distribution - CDF':
    ax.set_ylabel('$f_{n}(t)$', fontsize=15)
ax.set_title(title, fontsize=15)
ax.legend(loc='best', frameon=True)
ax.grid(True)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

title = 'F distribution - PDF'
# title = 'F distribution - CDF'
fig, ax = plt.subplots()
t = np.linspace(0.01, 10, 1000)
parameters_m = [2, 20]
parameters_n = [2, 20]

for m in parameters_m:
    for n in parameters_n:
        if title == 'F distribution - PDF':
            ax.plot(t, f.pdf(t,m,n), linewidth=3, label='$m=$'+str(m)+'$, n=$'+str(n))
        if title == 'F distribution - CDF':
            ax.plot(t, f.cdf(t,m,n), linewidth=3, label='$m=$'+str(m)+'$, n=$'+str(n))

ax.set_xlabel('$t$', fontsize=15)
if title == 'F distribution - PDF':
    ax.set_ylabel('$f_{m,n}(t)$', fontsize=15)
if title == 'F distribution - CDF':
    ax.set_ylabel('$f_{m,n}(t)$', fontsize=15)
ax.set_title(title, fontsize=15)
ax.legend(loc='best', frameon=True)
ax.grid(True)
```
