```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
from mpl_toolkits.mplot3d import Axes3D

mu_x = 0
mu_y = 0
var_x = 1
var_y = 1
cov_xy = 0
# Example from tutorial corresponds to:
# mu_x = mu_y = 0
# var_x = var_y = 1
# cov_xy = 0
# -> Projections on x- and y-axes are independent with standard normal distribution

x = np.linspace(-5,5,1000)
y = np.linspace(-5,5,1000)
X, Y = np.meshgrid(x,y)
pos = np.empty(X.shape + (2,))
pos[:, :, 0] = X; pos[:, :, 1] = Y
rv = multivariate_normal([mu_x, mu_y], [[var_x, cov_xy], [cov_xy, var_y]])

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(X, Y, rv.pdf(pos), cmap='viridis', linewidth=0.1, edgecolor='black')
ax.set_xlabel('$x$', fontsize=15)
ax.set_ylabel('$y$', fontsize=15)
ax.set_zlabel('$f(x,y)$', fontsize=15)
ax.set_title('Bivariate normal distribution - PDF', fontsize=15)
fig.set_figwidth(10)
fig.set_figheight(10)
plt.show()
```
