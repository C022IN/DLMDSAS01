```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon
from scipy.stats import gamma

theta0 = 10
theta1 = 9.5
n = 100
alpha = 0.05
x = expon.rvs(scale=theta1, size=n)
y = np.zeros(n)
t = np.mean(x)
c1 = gamma.ppf(alpha, n, scale=theta0)
c2 = c1/n

fig, ax = plt.subplots()
plt.axhline(y=0, color='0.2', linestyle='-')
ax.scatter(x, y, s=100)
ax.get_yaxis().set_visible(False)
ax.set_xlim(-0.2,np.max(x)+0.5)
ax.set_ylim(-0.15,0.15)
ax.set_title('Sample of size $n$ from $Exp(1/\\theta)$', fontsize=15)
fig.set_figwidth(8)
fig.set_figheight(3)
plt.show()

fig, ax = plt.subplots()
plt.axhline(y=0, color='0.2', linestyle='-')
ax.scatter(x, y, s=100)
ax.vlines(t, -0.1, 0.1, colors='b', label='$\\frac{1}{n} \\sum_{i=1}^n X_i$')
ax.vlines(c2, -0.1, 0.1, colors='r', label='$\\frac{1}{n} \\cdot [\\alpha$-quantile of $\Gamma(n,1/\\theta)]$')
ax.get_yaxis().set_visible(False)
ax.set_xlim(-0.2,np.max(x)+0.5)
ax.set_ylim(-0.15,0.25)
ax.set_title('Sample of size $n$ from $Exp(1/\\theta)$', fontsize=15)
ax.legend(loc='upper center', ncol=3, fontsize='large', frameon=True)
fig.set_figwidth(8)
fig.set_figheight(4)
plt.show()
```
