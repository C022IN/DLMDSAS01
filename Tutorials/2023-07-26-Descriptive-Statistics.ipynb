## Dataset and Random Variables
* A description can be done for two things in a value-space (often $\mathbb R$ or ${\mathbb R}^n$)
	* a dataset: a finite sequence of values (may have repetitions, hence no set): $x_1, ..., x_N \in {\mathbb R}^n$
	*  a random variable: the set of possible values and its probability (continuous or discrete finite or discrete infinite): $ X: {\mathcal S} \rightarrow {\mathbb R}^n$
* Descriptions can carry the same name or appearance
* Calculations are completely different
* but try to imitate: expectation of similarity, convergence criteria

## Describing by Values: Centrality: Mean (Expection)

* Idea: find value where things "most happen": a single value
* Mean of a dataset $(x_1, \ldots, x_n)$:
  $$ \overline x = \frac{1}{n} \sum_{i=1}^n x_i $$
* Note: if a value occurs three times it will be counted three times...
* Expectation of a random discrete variable $X:{\mathcal S} \rightarrow \mathbb R$ with pmf $f_X$: similar:
  $$ E(X) = \sum_{x} x\cdot f_X(x)$$
* Generalize to continuous random variable  $X:{\mathcal S} \rightarrow \mathbb R$ with pdf $f_X$: similar:
   $$ E(X) = \int_{-\infty}^{\infty} x\cdot f_X(x)dx$$
* Note: does not always exist!

* How to calculate? Mostly a theoretical question
* But note: all random variable descriptions document how to calculate the expectation.
* Exercise on a random dataset: calc the mean, show on plot
* Classical problem: mean gets influenced by outliers

## Describing by Values: Centrality: Mode
* Idea: find value with biggest "count": a single value
* dataset: value which occurs most often (top of histogram)
* discrete variable: max of pmf (top of bar chart)
* continuous variable: max of pdf (top of line graph)

## Describing by Values: Centrality: Median (single dim)
* Idea: find middle of data: a single value
* Median of dataset:
    * order the points by value
    * take value with half of the points left and half on the right
    * ... or midpoint of the two middle points
* Median of random variables: same idea but precise definitions vary

## Quantiles (single dim)
* Generalise idea: find the value at which x % is there
* e.g. 25-percentile: value for which a quarter of the values are there
* exercise: python to calculate it by a simple loop
* Note: not always well defined with a discrete distribution



## Describing by Values: Spread: Variance
* Idea: measure the mean of the distance to the mean
* Variance of a dataset:
  $$ s^2 = \frac{1}{N} \sum_{i=1}^n (x_i-\overline x)^2 $$
* Often one takes the root of this sum of squares:
  $$ s = \sqrt{\frac{1}{N} \sum_{i=1}^n (x_i-\overline x)^2} $$
* Expectation of a random discrete variable $X:{\mathcal S} \rightarrow \mathbb R$ with pmf $f_X$: similar:
    $$ Var(X) = \sum_{x} (x-E(X))^2\cdot f_X(x)$$

* Generalize to continuous random variable  $X:{\mathcal S} \rightarrow \mathbb R$ with pdf $f_X$: similar:
   $$ Var(X) = \int_{-\infty}^{\infty} (x-E(X))^2\cdot f_X(x)$$
* Warning: $Var(X)$ is NOT with the same unit ($unit^2$)

## Describing by Values: Symmetry
* Skewness: are the values more to right? (a number)
  $$ E \left[ \left ( \frac{X-E(X)}{s}\right)^3 \right ]$$
* Kurtosis: are the edges more think?
  $$ E \left[ \left ( \frac{X-E(X)}{s}\right)^4 \right ]$$
* one can calculate many more such "moments"
