```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

title = 'Beta distribution - PDF'
# title = 'Beta distribution - CDF'
fig, ax = plt.subplots()
t = np.linspace(0.01, 0.99, 1000)
parameters_a = [0.5, 2.0]
parameters_b = [0.5, 2.0]

for a in parameters_a:
    for b in parameters_b:
        if title == 'Beta distribution - PDF':
            ax.plot(t, beta.pdf(t,a,b), linewidth=3, label='$a=$'+str(a)+'$, b=$'+str(b))
        if title == 'Beta distribution - CDF':
            ax.plot(t, beta.cdf(t,a,b), linewidth=3, label='$a=$'+str(a)+'$, b=$'+str(b))

ax.set_xlabel('$t$', fontsize=15)
if title == 'Beta distribution - PDF':
    ax.set_ylabel('$f_{a,b}(t)$', fontsize=15)
if title == 'Beta distribution - CDF':
    ax.set_ylabel('$f_{a,b}(t)$', fontsize=15)
ax.set_title(title, fontsize=15)
ax.legend(loc='best', frameon=True)
ax.grid(True)
```
